<h4> Random Forest Classifier using digits dataset in sklearn </h4>

```python
import pandas as pd
from sklearn.datasets import load_digits
digits=load_digits()
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df=pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(['target'],axis='columns'),df.target,test_size=.2)
```

```python
len(X_test)
```

```python
model=RandomForestClassifier(n_estimators=45)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

<h4> to find the n_estimators for max accuracy </h4>

```python
scores=[]
n_estimators=range(1,51)
for i in n_estimators :
    model=RandomForestClassifier(n_estimators=i)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test,y_test))
    print('score:{}, n_estimator:{}'.format(scores[i-1],i))

plt.plot(n_estimators,scores)
plt.xlabel('n_estimators')
plt.ylabel('testing accuracy')
```
